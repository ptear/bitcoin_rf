```python
df_hr = pd.read_csv("BTCUSD_1hr.csv", parse_dates=[1])  # https://www.kaggle.com/prasoonkottarathil/btcinusd
```

```python
df_hr.head()
```

```python
df_hr.dtypes
```

```python
df_hr = df_hr.drop(0)
```

```python
df_hr.head()
```

```python
df_hr.tail()
```

```python
df_hr = df_hr[:-11]
```

```python
df_hr.tail()
```

```python
df_hr["day"] = df_hr["Date"].dt.date
df_hr.head()
```

```python
df_hr["hour"] = df_hr["Date"].dt.hour
df_hr.head()
```

```python
df_hr["returns"] = df_hr["Close"] - df_hr["Open"]
df_hr.head()
```

```python
df_hr_piv = df_hr.pivot(index="day", columns="hour", values="returns").reset_index()
df_hr_piv.head()
```

```python
df.head()
```

```python
df = df[::-1]
df.head()
```

```python
df_hr_piv.tail()
```

```python
df.tail()
```

```python
df = df[1:]
df.head()
```

```python
df = df[:-1]
df.tail()
```

```python
df_hr_piv.rename(columns = {'day':'Date'}, inplace = True)
```

```python
df_hr_piv.dtypes
```

```python
df_hr_piv['Date'] =  pd.to_datetime(df_hr_piv['Date'], infer_datetime_format=True)
```

```python
df_hr_piv.dtypes
```

```python
df_merged = df.merge(df_hr_piv, on="Date")
df_merged.head()
```

```python
df_merged.shape
```

```python
df_merged.tail()
```

```python
for hour in range(24):
    df_merged[f"{hour}:00"] = df_merged[hour].shift(1)
    
df_merged.tail()
```

```python
df_merged.head()
```

```python
df_new = df_merged.loc[:, "0:00":"23:00"]
df_new.head()
```

```python
df_new.insert(0, "Date", df_merged["Date"])
df_new.insert(1, "Returns", df_merged["returns"])
df_new.head()
```

```python
df_new = df_new.drop(0)
df_new.head()
```

```python
df_new.tail()
```

```python
df_new = df_new.round(5)
```

```python
df_new.to_csv("btc_returns.csv", index=False)
```
